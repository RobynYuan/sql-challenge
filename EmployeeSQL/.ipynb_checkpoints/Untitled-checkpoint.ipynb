```python

import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress
```

```python
salaries= pd.read_csv('Data/salaries.csv')
titles=pd.read_csv('Data/titles.csv')
employees=pd.read_csv('Data/employees.csv')
salaries.head()
```

```python
# Create a histogram to visualize the most common salary ranges for employees
salaries.hist(column='salary')






```

```python
# Create a bar chart of average salary by title.
salaries_employees= pd.merge(salaries,employees, on="emp_no", how="left")
salaries_employees=salaries_employees.rename({'emp_title_id':'title_id'}, axis='columns')
salaries_employees
```

```python
titles_salaries_employees= pd.merge(salaries_employees,titles, on="title_id", how="left")
titles_salaries_employees
```

```python
Average_salaries=titles_salaries_employees.groupby('title').mean()
Average_salaries

```

```python
plt.bar(Average_salaries.index.values,Average_salaries.salary)

plt.xticks(rotation=45)
plt.show()
```
